```python
import os
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
#read reviews
path = '../../../data/new-york-city-airbnb-open-data/'
reviews_csv = os.path.join(path,'reviews.csv')

reviews = pd.read_csv(reviews_csv)

#read listings
listings_csv = os.path.join(path,'listings.csv')

listings = pd.read_csv(listings_csv)
```

```python
#number_of_reviews_ltm
filt = listings['number_of_reviews_ltm'] < 100

df = listings['number_of_reviews_ltm']
bins = 10
mean = df.mean()
median = df.median()

plt.hist(df, bins=bins, edgecolor='black', log=True)
plt.axvline(mean, color='black', label='Mean')
plt.axvline(median, color='red', label='Median')

plt.legend()
plt.show()
```

```python
#'number_of_reviews', 'number_of_reviews_ltm', 'review_scores_rating',
df1 = listings['number_of_reviews_ltm']
df2 = listings['number_of_reviews']

print('correlation = {}'.format(df1.corr(df2)))

#chart
plt.scatter(df1, df2)
plt.xlabel('df1')
plt.ylabel('df2')
plt.show()

#correlation almost 0.76, i think these are similar features
```
